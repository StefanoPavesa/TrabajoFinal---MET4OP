```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import numpy as np
```

```python
censo = pd.read_excel('Censo_caba.xlsx')
```

```python
circuitos = gpd.read_file('circuitos-electorales.csv')
```

```python
circuitos
```

```python
mapaNBI = circuitos.merge(censo, on='COMUNA', how='left')
```

```python
mapaNBI
```

```python
circuitos['COMUNA'].unique()
```

```python
censo['COMUNA'].unique()
```

```python
circuitos['COMUNA'] = circuitos['COMUNA'].astype(int)
```

```python
circuitos['COMUNA'].unique()
```

```python
mapaNBI1 = circuitos.merge(censo, on='COMUNA', how='left')
```

```python
mapaNBI1
```

```python
mapaNBI1.plot()
```

```python
nbi_agrupado = mapaNBI1.groupby('COMUNA')['PoblacionNBI'].sum().reset_index()
```

```python
nbi_agrupado
```

```python
nbi_agrupado = mapaNBI1.groupby(['COMUNA']).agg({
    'WKT': 'first',
    'CIRCUITO_N': 'first',
    'PERIMETRO': 'first',
    'AREA': 'first',
    'geometry': 'first',
    'PoblacionNBI': 'sum'
}).reset_index()
```

```python
nbi_agrupado
```

```python
gdf_nbi_agrupado = gpd.GeoDataFrame(nbi_agrupado, geometry='geometry')

gdf_nbi_agrupado.plot(column='PoblacionNBI', cmap='viridis', legend=True)
plt.title('Población NBI por Comuna')
plt.show()
```

```python
gdf_nbi_agrupado = gpd.GeoDataFrame(nbi_agrupado, geometry='geometry')

gdf_nbi_agrupado.plot(column='PoblacionNBI', cmap='Blues', legend=True)
plt.title('Población NBI por Comuna')
plt.show()
```
