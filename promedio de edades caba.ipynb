```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
archivo_xls_2 = 'promedio_edad_caba (1).xlsx'
```

```python
promedio_edad = pd.read_excel(archivo_xls_2)
```

```python
print (promedio_edad)
```

```python
promedio_edad = promedio_edad.dropna(how='all')
```

```python
df = gpd.read_file('circuitos-electorales.csv')
```

```python
df.plot()
```

```python
df1 = df.rename(columns={'COMUNA': 'Comuna'})
```

```python
df1['Comuna'].unique()
```

```python
df1['Comuna'] = df1['Comuna'].replace({'1':"Comuna 1", '2':'Comuna 2', '3':'Comuna 3', '4':'Comuna 4', '5':'Comuna 5', '6':'Comuna 6', '7':'Comuna 7', '8':'Comuna 8', '9':'Comuna 9', '10':'Comuna 10', '11':'Comuna 11', '12':'Comuna 12', '13':'Comuna 13', '14':'Comuna 14', '15':'Comuna 15'})
```

```python
df1
```

```python
resultado = pd.merge(df1, promedio_edad, on = 'Comuna')
```

```python
resultado
```

```python
resultado_pandas = gpd.GeoDataFrame(resultado, geometry='geometry')
```

```python
resultado_pandas.plot(column='Total poblacion', cmap='turbo', legend=True, legend_kwds={'label': "Promedio de edad general"}, figsize=(10, 10))
plt.title('Mapa de calor: Promedio de edad por comuna')
plt.show()
```

```python
resultado_pandas.plot(column='Mujer/Femenino', cmap='Set1', legend=True, legend_kwds={'label': "Promedio de edad femenino"}, figsize=(10, 10))
plt.title('Mapa de calor: Promedio de edad de mujeres por comuna')
plt.show()
```

```python
resultado_pandas.plot(column='Varón/Masculino', cmap='Dark2', legend=True, legend_kwds={'label': "Promedio de edad masculino"}, figsize=(10, 10))
plt.title('Mapa de calor: Promedio de edad de varones por comuna')
plt.show()
```
