```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
df = gpd.read_file('circuitos-electorales.csv')
```

```python
df.plot()
```

```python
df1 = df.rename(columns={'COMUNA': 'seccion_nombre'})
```

```python
df1
```

```python
df1['seccion_nombre'].unique()
```

```python
df1['seccion_nombre'] = df1['seccion_nombre'].replace({'1':"Comuna 01", '2':'Comuna 02', '3':'Comuna 03', '4':'Comuna 04', '5':'Comuna 05', '6':'Comuna 06', '7':'Comuna 07', '8':'Comuna 08', '9':'Comuna 09', '10':'Comuna 10', '11':'Comuna 11', '12':'Comuna 12', '13':'Comuna 13', '14':'Comuna 14', '15':'Comuna 15'})
```

```python
df1
```

```python
pd.set_option('mode.chained_assignment', None)
```

```python
df2 = pd.read_csv('ResultadosElectorales_2023.csv', low_memory=False)
```

```python
resultado = pd.merge(df1, df2, on='seccion_nombre', how='inner')
```

```python
resultado
```

```python
resultados2 = resultado[(resultado['cargo_nombre'] == 'PRESIDENTE Y VICE') & (resultado['agrupacion_nombre'] == 'UNION POR LA PATRIA')]
```

```python
resultados2
```

```python
resultados_agrupados = resultados2.groupby(['WKT', 'ID', 'CIRCUITO_N', 'seccion_nombre', 'BARRIO', 'PERIMETRO', 'AREA', 'agrupacion_nombre']).agg({'votos_cantidad': 'sum'}).reset_index()
```

```python
resultados_agrupados
```

```python
import geopandas as gpd
from shapely import wkt
import matplotlib.pyplot as plt

# Supongamos que resultados_agrupados es tu DataFrame agrupado

# Convertir la columna 'WKT' a geometrÃ­as
resultados_agrupados['geometry'] = resultados_agrupados['WKT'].apply(wkt.loads)

# Crear un GeoDataFrame
gdf = gpd.GeoDataFrame(resultados_agrupados, geometry='geometry')

# Plotear el mapa de calor
gdf.plot(column='votos_cantidad', cmap='viridis', legend=True, legend_kwds={'label': "Votos Cantidad"}, figsize=(10, 10))
plt.title('Mapa de Calor de Votos Cantidad por Ãrea GeogrÃ¡fica')
plt.show()
```

```python
import geopandas as gpd
from shapely import wkt
import matplotlib.pyplot as plt

# Supongamos que resultados_agrupados es tu DataFrame agrupado

# Convertir la columna 'WKT' a geometrÃ­as
resultados_agrupados['geometry'] = resultados_agrupados['WKT'].apply(wkt.loads)

# Crear un GeoDataFrame
gdf = gpd.GeoDataFrame(resultados_agrupados, geometry='geometry')

# Definir un rango para la escala de colores (ajusta segÃºn tus datos)
vmin = resultados_agrupados['votos_cantidad'].min()
vmax = resultados_agrupados['votos_cantidad'].max()

# Plotear el mapa de calor con paleta de colores azules mÃ¡s fuertes y suaves
fig, ax = plt.subplots(figsize=(10, 10))
gdf.plot(column='votos_cantidad', cmap='Blues', legend=True, legend_kwds={'label': "Votos Cantidad"}, ax=ax, vmin=vmin, vmax=vmax)
plt.title('Mapa de Calor de Votos Cantidad por Ãrea GeogrÃ¡fica (Azules Fuertes y Suaves)')

# Eliminar etiquetas de los ejes x e y
ax.set_xticks([])
ax.set_yticks([])

plt.show()
```
