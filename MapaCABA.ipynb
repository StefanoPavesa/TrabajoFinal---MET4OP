```python
import pandas as pd
import matplotlib.pyplot as plt
import geopandas as gpd

%matplotlib inline
plt.style.use('bmh')
```

```python
comunas = gpd.read_file('circuito_01.shp')
```

```python
fig, ax = plt.subplots(figsize=(20, 20))

comunas.plot(linewidth=3, edgecolor='black', facecolor='None', ax=ax)

ax.set_title('Circuitos electorales - CABA')

ax.set_facecolor('White')
plt.axis('off')
```
